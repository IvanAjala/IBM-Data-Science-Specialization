```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create individual maps for each proximity type
def create_proximity_type_map(proximity_type):
    """Create a map focused on one specific proximity type"""
    proximity_map = folium.Map(location=[30, -100], zoom_start=5)
    
    style = proximity_styles[proximity_type]
    
    for site_name, coords in proximity_data.items():
        launch_coords = coords['launch_site']
        target_coords = coords[proximity_type]
        
        distance = calculate_distance(
            launch_coords[0], launch_coords[1],
            target_coords[0], target_coords[1]
        )
        
        # Add PolyLine
        folium.PolyLine(
            locations=[launch_coords, target_coords],
            weight=style['weight'],
            color=style['color'],
            opacity=0.8,
            dash_array=style['dash_array'],
            popup=f"{distance:.2f} km"
        ).add_to(proximity_map)
        
        # Add markers
        folium.Marker(
            launch_coords,
            popup=f"{site_name}",
            icon=folium.Icon(color='red', icon='rocket')
        ).add_to(proximity_map)
        
        folium.Marker(
            target_coords,
            popup=f"{proximity_type.title()}",
            icon=folium.Icon(color=style['color'], icon=style['icon'])
        ).add_to(proximity_map)
    
    return proximity_map

# Create individual maps for each proximity type
print("\nCreating individual proximity maps...")

# Uncomment to display individual maps:
# coastline_map = create_proximity_type_map('coastline')
# coastline_map
# 
# city_map = create_proximity_type_map('city')  
# city_map
# 
# highway_map = create_proximity_type_map('highway')
# highway_map
```
